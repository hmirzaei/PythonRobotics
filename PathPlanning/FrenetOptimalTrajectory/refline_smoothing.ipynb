```python
%matplotlib notebook
import time
import pickle
import math
from matplotlib import patches
import matplotlib.pyplot as plt
import numpy as np
from numpy import pi, cos, sin, tan, hypot, arctan, arctan2

# spg way points
(wx, wy) = pickle.load(open('ref_line4.pkl', 'rb'))

def draw_vehicle(fig, x, y, theta, delta, L , w, h, ww, wh, area, axle):
    if axle == 'front':
        x -= L/2 * cos(theta)
        y -= L/2 * sin(theta)
    elif axle == 'rear':
        x += L/2 * cos(theta)
        y += L/2 * sin(theta)
        
    ax = fig.gca()
    ax.clear()
    ax.plot(x,y,'.')
    rect = patches.Rectangle(
        (x, y), w/2, h/2,theta * 180 / np.pi, fill=False)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x, y), -w/2, h/2, theta * 180 / np.pi, fill=False)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x, y), w/2, -h/2, theta * 180 / np.pi, fill=False)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x, y), -w/2, -h/2, theta * 180 / np.pi, fill=False)
    ax.add_patch(rect)

    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta+arctan(h/L)), y-hypot(L/2,h/2)*sin(theta+arctan(h/L))), ww/2, wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta+arctan(h/L)), y-hypot(L/2,h/2)*sin(theta+arctan(h/L))), -ww/2, wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta+arctan(h/L)), y-hypot(L/2,h/2)*sin(theta+arctan(h/L))), ww/2, -wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta+arctan(h/L)), y-hypot(L/2,h/2)*sin(theta+arctan(h/L))), -ww/2, -wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)

    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta-arctan(h/L)), y-hypot(L/2,h/2)*sin(theta-arctan(h/L))), ww/2, wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta-arctan(h/L)), y-hypot(L/2,h/2)*sin(theta-arctan(h/L))), -ww/2, wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta-arctan(h/L)), y-hypot(L/2,h/2)*sin(theta-arctan(h/L))), ww/2, -wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x-hypot(L/2,h/2)*cos(theta-arctan(h/L)), y-hypot(L/2,h/2)*sin(theta-arctan(h/L))), -ww/2, -wh/2, theta * 180 / np.pi, fill=True)
    ax.add_patch(rect)


    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta-arctan(h/L)), y+hypot(L/2,h/2)*sin(theta-arctan(h/L))), ww/2, wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta-arctan(h/L)), y+hypot(L/2,h/2)*sin(theta-arctan(h/L))), -ww/2, wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta-arctan(h/L)), y+hypot(L/2,h/2)*sin(theta-arctan(h/L))), ww/2, -wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta-arctan(h/L)), y+hypot(L/2,h/2)*sin(theta-arctan(h/L))), -ww/2, -wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)

    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta+arctan(h/L)), y+hypot(L/2,h/2)*sin(theta+arctan(h/L))), ww/2, wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta+arctan(h/L)), y+hypot(L/2,h/2)*sin(theta+arctan(h/L))), -ww/2, wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta+arctan(h/L)), y+hypot(L/2,h/2)*sin(theta+arctan(h/L))), ww/2, -wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)
    rect = patches.Rectangle(
        (x+hypot(L/2,h/2)*cos(theta+arctan(h/L)), y+hypot(L/2,h/2)*sin(theta+arctan(h/L))), -ww/2, -wh/2, (theta + delta) * 180 / np.pi, fill=True)
    ax.add_patch(rect)

    ax.grid()
    plt.xlim(0 - area, 0 + area)
    plt.ylim(0 - area, 0 + area)
```

```python
T = 35 # simulation time
L = 3.78  # (m) rear axle to front axle distance
max_steer_angle = 25 # (deg) maximum steering angle
fs = 10 # samlping frequency
v = 5 # velocity
w = 5.9  # (m) vehicle width
h = 2  # (m) vehicle height
ww = 1  # (m) wheel width (used in visualizaion only)
wh = 0.3  # (m) wheel height (used in visualizaion only)
area = 30  # (m) visualizaion area
ref_point = 'rear'
k_th = 0.05 # heading error gain
k_e = 0.05  # lateral error gain
radi_curv = 10000 # fixed radius of curvature used in read pose feedback tracking
path = np.array(list(zip(wx, wy))) # reference path recorded at SPG


delta_a_sm = [0]  # initial delta (steering - smothooted array) just to supress errors
delta_rate_a_sm = [0] # initial delta rate (steering rate - smothooted array) just to supress errors
```

```python
def run_sim(steering_acc_input=[], use_steering_acc_input=True): # if use_steering_acc_input input is False 'rear pose feedback' tracking will be run
    fig = plt.figure(figsize=(6,6))
    ax = fig.gca()
    ax.set_aspect('equal', adjustable='box')
    fig.show()
    fig.canvas.draw()

    theta = np.arctan2(wy[3]-wy[0], wx[3]-wx[0])

    path_a = np.array([[wx[0], wy[0]]])
    cnt = 0
    x = wx[0]
    y = wy[0]
    delta = delta_a_sm[0]
    steer_rate = delta_rate_a_sm[0]
    delta_a = []
    x_sm = []
    y_sm = []
    prev_i = 0
    for t in np.linspace(0, T, int(T * fs)): # time
        x_sm.append(x)
        y_sm.append(y)

        if use_steering_acc_input:
            steer_acc = steering_acc_input[cnt]
            steer_rate += steer_acc/fs
            delta += steer_rate/fs
        else:
            ######################################## rear pose feedback tracking
            path2 = path[prev_i:prev_i+300]
            i = np.argmin(np.linalg.norm(path2 - [x, y],axis=1)) # index of min distance point on path to ref point
            e = np.linalg.norm(path2 - [x, y],axis=1)[i]      # min distance to path
            if i < len(path2) - 1:
                x1, y1, x2, y2 = path2[i,0], path2[i,1], path2[i+1,0], path2[i+1,1]
            else:
                x1, y1, x2, y2 = path2[-2,0], path2[-2,1], path2[-1,0], path2[-1,1]
            prev_i = i + prev_i

            path_a_tang = [cos(theta), sin(theta)]
            path_t_tang = [x2-x1, y2-y1]
            path_t_tang /= np.linalg.norm(path_t_tang)

            d = (x-x1)*(y2-y1)-(y-y1)*(x2-x1)
            e = e * np.sign(d)

            th_e = np.arcsin(np.cross(path_a_tang, path_t_tang))
            omega = (v * (1/radi_curv) * cos(th_e))/(1 - (1/radi_curv) * e) - k_th * abs(v) * th_e - k_e * v * sin (th_e)/th_e * e
            delta = -arctan(L * omega / v)        

        delta = np.clip(delta,-max_steer_angle*pi/180,max_steer_angle*pi/180)    
        delta_a.append(delta)

        x += v*cos(theta)/fs
        y += v*sin(theta)/fs
        theta += v*tan(delta)/L/fs

        path_a = np.append(path_a,np.array([[x, y]]), axis = 0)
        cnt += 1
        if cnt%200 == 0 or cnt == int(T * fs) - 1: # control animation speed
            draw_vehicle(fig, x, y, theta, delta, L , w, h, ww, wh, area, ref_point)

            ax.plot(path[::1,0],path[::1,1],'.-')
            ax.plot(path_a[::1,0],path_a[::1,1],'*-')
            plt.title("delta:" + '%.2f' % (delta * 180 / pi) + "  t:" + '%.2f' % (t))
            plt.xlim(x - area, x + area)
            plt.ylim(y - area, y + area)

            fig.canvas.draw()  # draw
            #time.sleep(0.5)  # sleep
    return delta_a, (x_sm, y_sm)
```

```python
# run sim with reference tracking method (rear pose feedback)
delta_a, result_path = run_sim(use_steering_acc_input=False)

# calculate smoothen second derivative of steering command
window=np.hamming(1)
delta_a_sm=np.convolve(window/window.sum(),np.array(delta_a),mode='valid')
dalta_rate_a=np.diff(delta_a_sm)*fs
delta_rate_a_sm=np.convolve(window/window.sum(),np.array(dalta_rate_a),mode='valid')
steer_acc_a_sm = np.diff(delta_rate_a_sm)*fs
steer_acc_a_sm = np.r_[steer_acc_a_sm, [steer_acc_a_sm[-1]]]
steer_acc_a_sm = np.r_[steer_acc_a_sm, [steer_acc_a_sm[-1]]]
plt.figure()
plt.plot(steer_acc_a_sm)

# run sim with smoothen second derivative of steering command obtained by reference method
dummy = run_sim(steer_acc_a_sm)
```

```python
weight = 0.2
horizon = 50
def cost_fun(xx):
    delta = delta_a_sm[0]
    steer_rate = delta_rate_a_sm[0]
    x = wx[0]
    y = wy[0]
    theta = np.arctan2(wy[3]-wy[0], wx[3]-wx[0])
    fs = 10
    prev_i = 0
    e = 0
    for cnt in range(horizon): # time
        t = cnt / fs
        path2 = path[prev_i:prev_i+10]
        i = np.argmin(np.linalg.norm(path2 - [x, y],axis=1)) # index of min distance point on path to ref point
        e += weight * np.linalg.norm(path2 - [x, y],axis=1)[i] + (1-weight) * abs(xx[cnt])      # min distance to path
        prev_i = i + prev_i

        steer_acc = xx[cnt]
        steer_rate += steer_acc/fs
        delta += steer_rate/fs

        delta = np.clip(delta,-max_steer_angle*pi/180,max_steer_angle*pi/180)    

        x += v*cos(theta)/fs
        y += v*sin(theta)/fs
        theta += v*tan(delta)/L/fs
    return e
```

```python
from scipy.optimize import minimize, rosen, rosen_der

Nfeval = 1
def callbackF(Xi):
    global Nfeval
    if Nfeval%10 == 0:
        print('%d   %f'%(Nfeval, cost_fun(Xi)))
    Nfeval += 1

print('%s %s'%('Iter', 'f(X)'))


weight = 0.2
horizon = 50
opt_steer_acc = minimize(cost_fun, steer_acc_a_sm, method='BFGS', callback=callbackF, options=({'maxiter':100, 'disp':True}))
plt.figure()
plt.plot(opt_steer_acc.x)

for horizon in range(100,351,50):
    print('horizon = %d'%horizon)
    # use previous step solution as iniital guess
    Nfeval = 1
    opt_steer_acc = minimize(cost_fun, opt_steer_acc.x, method='BFGS', callback=callbackF, options=({'maxiter':100, 'disp':True}))
    plt.figure()
    plt.plot(opt_steer_acc.x)

# use previous step solution as iniital guess
weight = 0.8  # look for even more smoother path 
opt_steer_acc = minimize(cost_fun, opt_steer_acc.x, method='BFGS', callback=callbackF, options=({'maxiter':10, 'disp':True}))
plt.figure()
plt.plot(opt_steer_acc.x)
```
